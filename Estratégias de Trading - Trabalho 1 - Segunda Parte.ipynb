```python
# -*- coding: utf-8 -*-
"""
Created on Wed Feb 17 21:01:12 2021

@author: henri
"""


# # Escola de Economia de São Paulo
# ## Mestrado Profissional em Economia (MPE)
# ### Ênfase: Engenharia Financeira
# #### Disciplina: Estratégia de Trading
# ##### Professor: Alessandro Marques
# ##### Alunos: Erich Ferreira, Felipe Bernardo, Henrique Assis
# ##### Trabalho 1 - Segunda parte


import yfinance as yf
import pandas as pd
# import numpy as np
import statistics as stc
import scipy as sp
from statsmodels import robust
# import datetime
import math
# import os
# import sys


def get_bases():
    
    # Searching all the major indexes on the yahoo finance database
    df_list = pd.read_html('https://finance.yahoo.com/world-indices/')
    majorstockidx = df_list[0]
    all_tickers = majorstockidx['Symbol'].values.tolist()
    
    # Showing the user all the available indices
    print('')
    print('The major indices available for extraction at Yahoo Finance are the following:')
    print('')
    print(all_tickers)
    print('')
    
    # Define your directory to save the databases
    directory = "C:/Users/henri/OneDrive/Mestrado Engenharia Financeira/Segundo Semestre/Estratégia de Trading/Monitorias/Monitoria 1/databases/"
    
    # Index: Dow Jones Industrial Average
    # Start Date: 01/01/1980
    djia = yf.Ticker("^DJI")
    
    hist_djia_1d = djia.history(start='1980-01-01', end='2021-01-01', interval= '1d', prepost=True)
    hist_djia_1w = djia.history(start='1980-01-01', end='2021-01-01', interval= '1wk', prepost=True)
    hist_djia_1m = djia.history(start='1980-01-01', end='2021-01-01', interval= '1mo', prepost=True)
    
    hist_djia_1d.to_csv(directory + 'hist_djia_1d.csv')
    hist_djia_1w.to_csv(directory + 'hist_djia_1w.csv')
    hist_djia_1m.to_csv(directory + 'hist_djia_1m.csv')
    
    # Index: Standard & Poor's 500
    # Start Date: 01/01/1980
    gspc = yf.Ticker("^GSPC")
    
    hist_gspc_1d = gspc.history(start='1980-01-01', end='2021-01-01', interval= '1d', prepost=True)
    hist_gspc_1w = gspc.history(start='1980-01-01', end='2021-01-01', interval= '1wk', prepost=True)
    hist_gspc_1m = gspc.history(start='1980-01-01', end='2021-01-01', interval= '1mo', prepost=True)
    
    hist_gspc_1d.to_csv(directory + 'hist_gspc_1d.csv')
    hist_gspc_1w.to_csv(directory + 'hist_gspc_1w.csv')
    hist_gspc_1m.to_csv(directory + 'hist_gspc_1m.csv')
    
    # Index: NIKKEI
    # Start Date: 01/01/1980
    nikkei = yf.Ticker("^N225")
    hist_n225_1d = nikkei.history(start='1980-01-01', end='2021-01-01', interval= '1d', prepost=True)
    hist_n225_1w = nikkei.history(start='1980-01-01', end='2021-01-01', interval= '1wk', prepost=True)
    hist_n225_1m = nikkei.history(start='1980-01-01', end='2021-01-01', interval= '1mo', prepost=True)
    
    hist_n225_1d.to_csv(directory + 'hist_nikkei_1d.csv')
    hist_n225_1w.to_csv(directory + 'hist_nikkei_1w.csv')
    hist_n225_1m.to_csv(directory + 'hist_nikkei_1m.csv')
    
    # Index: New York Stock Exchange Composite
    # Start Date: 01/01/1980
    nya = yf.Ticker("^NYA")
    
    hist_nya_1d = nya.history(start='1980-01-01', end='2021-01-01', interval= '1d', prepost=True)
    hist_nya_1w = nya.history(start='1980-01-01', end='2021-01-01', interval= '1wk', prepost=True)
    hist_nya_1m = nya.history(start='1980-01-01', end='2021-01-01', interval= '1mo', prepost=True)
    
    hist_nya_1d.to_csv(directory + 'hist_nya_1d.csv')
    hist_nya_1w.to_csv(directory + 'hist_nya_1w.csv')
    hist_nya_1m.to_csv(directory + 'hist_nya_1m.csv')
    
    # Index: NASDAQ Composite Index
    # Start Date: 01/01/1980
    ixic = yf.Ticker("^IXIC")
    
    hist_ixic_1d = ixic.history(start='1980-01-01', end='2021-01-01', interval= '1d', prepost=True)
    hist_ixic_1w = ixic.history(start='1980-01-01', end='2021-01-01', interval= '1wk', prepost=True)
    hist_ixic_1m = ixic.history(start='1980-01-01', end='2021-01-01', interval= '1mo', prepost=True)
    
    hist_ixic_1d.to_csv(directory + 'hist_ixic_1d.csv')
    hist_ixic_1w.to_csv(directory + 'hist_ixic_1w.csv')
    hist_ixic_1m.to_csv(directory + 'hist_ixic_1m.csv')
    

def read_base():
    
    # Define your directory to save the databases
    directory = "C:/Users/henri/OneDrive/Mestrado Engenharia Financeira/Segundo Semestre/Estratégia de Trading/Monitorias/Monitoria 1/databases/"
    
    # The databases (.csv files) must be in the same directory as this .py!
    df_djia_1d = pd.read_csv(directory + 'hist_djia_1d.csv')
    df_djia_1w = pd.read_csv(directory + 'hist_djia_1w.csv')
    df_djia_1m = pd.read_csv(directory + 'hist_djia_1m.csv')
    
    df_gspc_1d = pd.read_csv(directory + 'hist_gspc_1d.csv')
    df_gspc_1w = pd.read_csv(directory + 'hist_gspc_1w.csv')
    df_gspc_1m = pd.read_csv(directory + 'hist_gspc_1m.csv')
    
    df_n225_1d = pd.read_csv(directory + 'hist_nikkei_1d.csv')
    df_n225_1w = pd.read_csv(directory + 'hist_nikkei_1w.csv')
    df_n225_1m = pd.read_csv(directory + 'hist_nikkei_1m.csv')
    
    df_nya_1d = pd.read_csv(directory + 'hist_nya_1d.csv')
    df_nya_1w = pd.read_csv(directory + 'hist_nya_1w.csv')
    df_nya_1m = pd.read_csv(directory + 'hist_nya_1m.csv')
    
    df_ixic_1d = pd.read_csv(directory + 'hist_ixic_1d.csv')
    df_ixic_1w = pd.read_csv(directory + 'hist_ixic_1w.csv')
    df_ixic_1m = pd.read_csv(directory + 'hist_ixic_1m.csv')
    
    return df_djia_1d, df_djia_1w, df_djia_1m, df_gspc_1d, df_gspc_1w, df_gspc_1m, df_n225_1d, df_n225_1w, df_n225_1m, df_nya_1d, df_nya_1w, df_nya_1m, df_ixic_1d, df_ixic_1w, df_ixic_1m


def month_to_year(dataframe_month):
    
    flagging = [0] * int(len(dataframe_month['Date']))
    dataframe_month['Flag'] = flagging
    dates_vector = dataframe_month['Date'].values.tolist()
    first_year = dates_vector[0]
    setting_year = int(first_year[0:4]) # Only works for this date format!
    setting_month = int(first_year[5:7]) # Only works for this date format!
    setting_day = int(first_year[8:10]) # Only works for this date format!
    df_year = pd.DataFrame(columns=('Date', 'Open', 'High', 'Low', 'Close', 'Volume', 'Dividends', 'Stock Splits', 'Flag'))
    j = 0
    
    for i in range(len(dataframe_month['Date'])):
        date_now = dates_vector[i]
        year_now = int(date_now[0:4])
        month_now = int(date_now[5:7])
        day_now = int(date_now[8:10])
        if (year_now == (setting_year + j)) & (month_now == setting_month) & (day_now == setting_day):
            dataframe_month['Flag'].iloc[[i]] = 1
            j += 1
        
    df_year = pd.concat([df_year, dataframe_month[dataframe_month['Flag'] == 1]])
    
    return df_year


def calc_logreturn(vector_values):
    
    log_return = []
    
    for i in range(len(vector_values)-1):
        log_return.append(math.log(vector_values[i+1]/vector_values[i]))
    
    return log_return


def calc_median(lr):
    
    # Using a pre-programmed function
    median1 = stc.median(lr)
    
    # Programming own function
    lr.sort()
    tam = len(lr)
    idx = tam / 2
    
    if (idx - math.floor(idx)) != 0:
        idx1 = math.floor(idx)
        idx2 = math.floor(idx) + 1
        median2 = (lr[idx1] + lr[idx2]) / 2
        
    else:
        idx1 = int(idx)
        median2 = lr[idx1]
    
    return [median1, median2]


def calc_mean(lr):
    
    # Using a pre-programmed function
    mean1 = stc.mean(lr)
    
    # Programming own function
    T = len(lr)
    Sum = 0
    
    for i in lr:
        Sum += i
        
    mean2 = Sum / T
    
    return [mean1, mean2]


def calc_stddev(lr):
    
    # Using a pre-programmed function
    stddev1 = stc.stdev(lr)
    
    # Using own function
    T = len(lr)
    Sum = 0
    
    for i in lr:
        Sum += i
        
    mean = Sum / T
    
    diff = []
    sum_diff = 0
    
    for i in range(len(lr)):
        diff.append(lr[i] - mean)
    
    for i in range(len(lr)):
        sum_diff += diff[i] ** 2
    
    stddev2 = (sum_diff / (len(lr) - 1)) ** 0.5
    
    return [stddev1, stddev2]


def calc_max(lr):
    
    # Using a pre-programmed function
    maximum1 = max(lr)
    
    # Using own function
    maximum2 = 0
    for i in range(len(lr)):
        if lr[i] > maximum2:
            maximum2 = lr[i]
    
    return [maximum1, maximum2]


def calc_min(lr):
    
    # Using a pre-programmed function
    minimum1 = min(lr)
    
    # Using own function
    minimum2 = 100
    for i in range(len(lr)):
        if lr[i] < minimum2:
            minimum2 = lr[i]
    
    return [minimum1, minimum2]


def calc_skewness(lr):
    
    # Using a pre-programmed function
    skewness1 = sp.stats.skew(lr)
    
    # Using own function
    T = len(lr)
    Sum = 0
    
    for i in lr:
        Sum += i
        
    mean = Sum / T
    
    diff = []
    sum_diff = 0
    
    for i in range(len(lr)):
        diff.append(lr[i] - mean)
    
    for i in range(len(lr)):
        sum_diff += diff[i] ** 2
    
    stddev = (sum_diff / (len(lr) - 1)) ** 0.5
    
    accum = 0
    
    for i in range(len(lr)):
        accum += (lr[i] - mean) ** 3
        
    skewness2 = accum * T / ((T - 1) * (T - 2) * stddev**3)
    
    return [skewness1, skewness2]


def calc_kurtosis(lr):
    
    # Using a pre-programmed function
    kurtosis1 = sp.stats.kurtosis(lr)
    
    # Using own fucntion
    T = len(lr)
    Sum = 0
    
    for i in lr:
        Sum += i
        
    mean = Sum / T
    
    diff = []
    sum_diff = 0
    
    for i in range(len(lr)):
        diff.append(lr[i] - mean)
    
    for i in range(len(lr)):
        sum_diff += diff[i] ** 2
    
    var = sum_diff / (len(lr) - 1)
    
    accum = 0
    
    for i in range(len(lr)):
        accum += (lr[i] - mean) ** 4
        
    kurtosis2 = (((T + 1) * T) / ((T - 1) * (T - 2) * (T - 3))) * (accum / var) - 3 * (((T - 1) ** 2) / (T - 2) * (T - 3))
    
    return [kurtosis1, kurtosis2]


def calc_mad(lr):
    
    # Using a pre-programmed function
    mad1 = robust.mad(lr)
    
    # Using own function
    T = len(lr)
    Sum = 0
    
    for i in lr:
        Sum += i
        
    mean = Sum / T
    
    sum_diff = 0
    
    for i in range(len(lr)):
        sum_diff += abs(lr[i] - mean)
        
    mad2 = sum_diff / T
    
    return [mad1, mad2]

def calc_iqr(lr):
    
    # Using a pre-programmed function
    iqr1 = sp.stats.iqr(lr)
    
    # Using own function
    lr.sort()
    tam = len(lr)
    idx4 = tam / 4
    first_quartile = 0
    
    if (idx4 - math.floor(idx4)) != 0:
        idx4_1 = math.floor(idx4)
        idx4_2 = math.floor(idx4) + 1
        first_quartile = (lr[idx4_1] + lr[idx4_2]) / 2
        
    else:
        idx4_1 = int(idx4)
        first_quartile = lr[idx4_1]
        
    third_quartile = 0
    idx3 = tam * (3 / 4)
    
    if (idx3 - math.floor(idx3)) != 0:
        idx3_1 = math.floor(idx3)
        idx3_2 = math.floor(idx3) + 1
        third_quartile = (lr[idx3_1] + lr[idx3_2]) / 2
        
    else:
        idx3_1 = int(idx3)
        third_quartile = lr[idx3_1]
        
    iqr2 = third_quartile - first_quartile
    
    return [iqr1, iqr2]


def normality_test(lr):
    
    # Using a pre-programmed function - Jarque-Bera Test
    # This test is better fitted for large samples (N > 5000)
    jb_test1 = sp.stats.jarque_bera(lr)
    
    # Using a pre-programmed function - Shapiro-Wilk Test
    # This test is better fitted for small samples (N < 5000)
    shapiro_test1 = sp.stats.shapiro(lr)
    
    # Using a pre-programmed function - Anderson-Darling Test
    # This test can be used for any distribution, the null hypothesis is that a sample is drawn from a particular distribution
    # anderson_test1 = sp.stats.anderson(lr, dist='norm')
    
    # Using a pre-programmed function - Kolmogorov-Smirnov Test
    # This test can be used for any distribution, the null hypothesis is that a sample is drawn from a particular distribution
    ks_test1 = sp.stats.kstest(lr, 'norm')
    
    # test1 = np.zeros((4,2), dtype=float)
    # test1[0,:] = jb_test1
    # test1[1,:] = shapiro_test1
    # test1[2,:] = anderson_test1
    # test1[3,:] = ks_test1
    
    test1 = [jb_test1, shapiro_test1, ks_test1]
    
    return test1
    

def construct_dataframe(median, mean, stddev, maxi, mini, skewness, kurtosis, mad, iqr, jb, shapiro, ks):
    
    # Setting up the dataframe header and columns
    head = ['Index', 'Sample Interval', 'Programmed Median', 'Implemented Median',
            'Programmed Mean', 'Implemented Mean', 'Programmed Standard Deviation',
            'Implemented Standard Deviation', 'Programmed Maximum', 'Implemented Maximum',
            'Programmed Minimum', 'Implemented Minimum', 'Programmed Skewness',
            'Implemented Skewness', 'Programmed Kurtosis', 'Implemented Kurtosis',
            'Programmed MAD', 'Implemented MAD', 'Programmed IQR', 'Implemented IQR',
            'Jarque-Bera Test Statistic', 'Jarque-Bera p-Value',
            'Shapiro-Wilk Test Statistic', 'Shapiro-Wilk p-Value',
            'Kolmogorov-Smirnov Test Statistic', 'Kolmogorov-Smirnov p-Value']
    column_1 = ['Dow Jones', 'Dow Jones', 'Dow Jones', 'Dow Jones', 'S&P 500',
                'S&P 500', 'S&P 500', 'S&P 500', 'Nikkei', 'Nikkei', 'Nikkei',
                'Nikkei', 'NYSE', 'NYSE', 'NYSE', 'NYSE', 'NASDAQ', 'NASDAQ',
                'NASDAQ', 'NASDAQ']
    column_2 = ['Daily', 'Weekly', 'Monthly', 'Yearly', 'Daily', 'Weekly', 'Monthly',
                'Yearly', 'Daily', 'Weekly', 'Monthly', 'Yearly', 'Daily', 'Weekly',
                'Monthly', 'Yearly', 'Daily', 'Weekly', 'Monthly', 'Yearly']
    
    # Building the dataframe
    df = pd.DataFrame(columns=head)
    df['Index'] = column_1
    df['Sample Interval'] = column_2
    df['Programmed Median'] = median[0]
    df['Implemented Median'] = median[1]
    df['Programmed Mean'] = mean[0]
    df['Implemented Mean'] = mean[1]
    df['Programmed Standard Deviation'] = stddev[0]
    df['Implemented Standard Deviation'] = stddev[1]
    df['Programmed Maximum'] = maxi[0]
    df['Implemented Maximum'] = maxi[1]
    df['Programmed Minimum'] = mini[0]
    df['Implemented Minimum'] = mini[1]
    df['Programmed Skewness'] = skewness[0]
    df['Implemented Skewness'] = skewness[1]
    df['Programmed Kurtosis'] = kurtosis[0]
    df['Implemented Kurtosis'] = kurtosis[1]
    df['Programmed MAD'] = mad[0]
    df['Implemented MAD'] = mad[1]
    df['Programmed IQR'] = iqr[0]
    df['Implemented IQR'] = iqr[1]
    df['Jarque-Bera Test Statistic'] = jb[0][0]
    df['Jarque-Bera p-Value'] = jb[0][1]
    df['Shapiro-Wilk Test Statistic'] = shapiro[0][0]
    df['Shapiro-Wilk p-Value'] = shapiro[0][1]
    df['Kolmogorov-Smirnov Test Statistic'] = ks[0][0]
    df['Kolmogorov-Smirnov p-Value'] = ks[0][1]
    
    return df
    
# Here the main code begins

print('')
print('This code will import from Yahoo Finance the historical data of five indexes: ')
print('')
print('1) Dow Jones Industrial Average (DJI);')
print("2) Standard and Poor's 500 (GSPC);")
print('3) Nikkei Stock Average (N225);')
print('4) New York Stock Exchange Composite (NYA);')
print('5) NASDAQ Composite Index (IXIC).')
print('')
print("This code is meant to bring all the index's data from 02-01-1980 to 31-12-2020.")
print('')
print('')
print('Do you want to access the databases from Yahoo Finance? (y/n)')
print('WARNING: If you alredy ran this code and alredy have those .csv databases saved, insert "n" for "no" !!!')
print('')
access_database = input('> ')

if access_database == 'y':
    get_bases()
    
# Searching for the daily, weekly and monthly databases from your computer
df_djia_1d, df_djia_1w, df_djia_1m, df_gspc_1d, df_gspc_1w, df_gspc_1m, df_n225_1d, df_n225_1w, df_n225_1m, df_nya_1d, df_nya_1w, df_nya_1m, df_ixic_1d, df_ixic_1w, df_ixic_1m = read_base()
  
# creating the yearly database  
df_djia_1y = month_to_year(df_djia_1m)
df_gspc_1y = month_to_year(df_gspc_1m)
df_n225_1y = month_to_year(df_n225_1m)
df_nya_1y = month_to_year(df_nya_1m)
df_ixic_1y = month_to_year(df_ixic_1m)

# Creating a list-like columns vector with the closing values of each index
list_djia_1d = df_djia_1d['Close'].values.tolist()
list_djia_1w = df_djia_1w['Close'].values.tolist()
list_djia_1m = df_djia_1m['Close'].values.tolist()
list_djia_1y = df_djia_1y['Close'].values.tolist()

list_gspc_1d = df_gspc_1d['Close'].values.tolist()
list_gspc_1w = df_gspc_1w['Close'].values.tolist()
list_gspc_1m = df_gspc_1m['Close'].values.tolist()
list_gspc_1y = df_gspc_1y['Close'].values.tolist()

list_n225_1d = df_n225_1d['Close'].values.tolist()
list_n225_1w = df_n225_1w['Close'].values.tolist()
list_n225_1m = df_n225_1m['Close'].values.tolist()
list_n225_1y = df_n225_1y['Close'].values.tolist()

list_nya_1d = df_nya_1d['Close'].values.tolist()
list_nya_1w = df_nya_1w['Close'].values.tolist()
list_nya_1m = df_nya_1m['Close'].values.tolist()
list_nya_1y = df_nya_1y['Close'].values.tolist()

list_ixic_1d = df_ixic_1d['Close'].values.tolist()
list_ixic_1w = df_ixic_1w['Close'].values.tolist()
list_ixic_1m = df_ixic_1m['Close'].values.tolist()
list_ixic_1y = df_ixic_1y['Close'].values.tolist()

# Calculating the log-returns for the daily, weekly, monthly and yearly databases
logreturn_djia_1d = calc_logreturn(list_djia_1d)
logreturn_djia_1w = calc_logreturn(list_djia_1w)
logreturn_djia_1m = calc_logreturn(list_djia_1m)
logreturn_djia_1y = calc_logreturn(list_djia_1y)

logreturn_gspc_1d = calc_logreturn(list_gspc_1d)
logreturn_gspc_1w = calc_logreturn(list_gspc_1w)
logreturn_gspc_1m = calc_logreturn(list_gspc_1m)
logreturn_gspc_1y = calc_logreturn(list_gspc_1y)
    
logreturn_n225_1d = calc_logreturn(list_n225_1d)
logreturn_n225_1w = calc_logreturn(list_n225_1w)
logreturn_n225_1m = calc_logreturn(list_n225_1m)
logreturn_n225_1y = calc_logreturn(list_n225_1y)

logreturn_nya_1d = calc_logreturn(list_nya_1d)
logreturn_nya_1w = calc_logreturn(list_nya_1w)
logreturn_nya_1m = calc_logreturn(list_nya_1m)
logreturn_nya_1y = calc_logreturn(list_nya_1y)
    
logreturn_ixic_1d = calc_logreturn(list_ixic_1d)
logreturn_ixic_1w = calc_logreturn(list_ixic_1w)
logreturn_ixic_1m = calc_logreturn(list_ixic_1m)
logreturn_ixic_1y = calc_logreturn(list_ixic_1y)

# Calculating the median for the daily, weekly, monthly and yearly databases
median_djia_1d = calc_median(logreturn_djia_1d)
median_djia_1w = calc_median(logreturn_djia_1w)
median_djia_1m = calc_median(logreturn_djia_1m)
median_djia_1y = calc_median(logreturn_djia_1y)

median_gspc_1d = calc_median(logreturn_gspc_1d)
median_gspc_1w = calc_median(logreturn_gspc_1w)
median_gspc_1m = calc_median(logreturn_gspc_1m)
median_gspc_1y = calc_median(logreturn_gspc_1y)

median_n225_1d = calc_median(logreturn_n225_1d)
median_n225_1w = calc_median(logreturn_n225_1w)
median_n225_1m = calc_median(logreturn_n225_1m)
median_n225_1y = calc_median(logreturn_n225_1y)

median_nya_1d = calc_median(logreturn_nya_1d)
median_nya_1w = calc_median(logreturn_nya_1w)
median_nya_1m = calc_median(logreturn_nya_1m)
median_nya_1y = calc_median(logreturn_nya_1y)

median_ixic_1d = calc_median(logreturn_ixic_1d)
median_ixic_1w = calc_median(logreturn_ixic_1w)
median_ixic_1m = calc_median(logreturn_ixic_1m)
median_ixic_1y = calc_median(logreturn_ixic_1y)

# Calculating the mean for the daily, weekly, monthly and yearly databases
mean_djia_1d = calc_mean(logreturn_djia_1d)
mean_djia_1w = calc_mean(logreturn_djia_1w)
mean_djia_1m = calc_mean(logreturn_djia_1m)
mean_djia_1y = calc_mean(logreturn_djia_1y)

mean_gspc_1d = calc_mean(logreturn_gspc_1d)
mean_gspc_1w = calc_mean(logreturn_gspc_1w)
mean_gspc_1m = calc_mean(logreturn_gspc_1m)
mean_gspc_1y = calc_mean(logreturn_gspc_1y)

mean_n225_1d = calc_mean(logreturn_n225_1d)
mean_n225_1w = calc_mean(logreturn_n225_1w)
mean_n225_1m = calc_mean(logreturn_n225_1m)
mean_n225_1y = calc_mean(logreturn_n225_1y)

mean_nya_1d = calc_mean(logreturn_nya_1d)
mean_nya_1w = calc_mean(logreturn_nya_1w)
mean_nya_1m = calc_mean(logreturn_nya_1m)
mean_nya_1y = calc_mean(logreturn_nya_1y)

mean_ixic_1d = calc_mean(logreturn_ixic_1d)
mean_ixic_1w = calc_mean(logreturn_ixic_1w)
mean_ixic_1m = calc_mean(logreturn_ixic_1m)
mean_ixic_1y = calc_mean(logreturn_ixic_1y)

# Calculating the standard deviation for the daily, weekly, monthly and yearly databases
stddev_djia_1d = calc_stddev(logreturn_djia_1d)
stddev_djia_1w = calc_stddev(logreturn_djia_1w)
stddev_djia_1m = calc_stddev(logreturn_djia_1m)
stddev_djia_1y = calc_stddev(logreturn_djia_1y)

stddev_gspc_1d = calc_stddev(logreturn_gspc_1d)
stddev_gspc_1w = calc_stddev(logreturn_gspc_1w)
stddev_gspc_1m = calc_stddev(logreturn_gspc_1m)
stddev_gspc_1y = calc_stddev(logreturn_gspc_1y)

stddev_n225_1d = calc_stddev(logreturn_n225_1d)
stddev_n225_1w = calc_stddev(logreturn_n225_1w)
stddev_n225_1m = calc_stddev(logreturn_n225_1m)
stddev_n225_1y = calc_stddev(logreturn_n225_1y)

stddev_nya_1d = calc_stddev(logreturn_nya_1d)
stddev_nya_1w = calc_stddev(logreturn_nya_1w)
stddev_nya_1m = calc_stddev(logreturn_nya_1m)
stddev_nya_1y = calc_stddev(logreturn_nya_1y)

stddev_ixic_1d = calc_stddev(logreturn_ixic_1d)
stddev_ixic_1w = calc_stddev(logreturn_ixic_1w)
stddev_ixic_1m = calc_stddev(logreturn_ixic_1m)
stddev_ixic_1y = calc_stddev(logreturn_ixic_1y)

# Calculating the maximum for the daily, weekly, monthly and yearly databases
max_djia_1d = calc_max(logreturn_djia_1d)
max_djia_1w = calc_max(logreturn_djia_1w)
max_djia_1m = calc_max(logreturn_djia_1m)
max_djia_1y = calc_max(logreturn_djia_1y)

max_gspc_1d = calc_max(logreturn_gspc_1d)
max_gspc_1w = calc_max(logreturn_gspc_1w)
max_gspc_1m = calc_max(logreturn_gspc_1m)
max_gspc_1y = calc_max(logreturn_gspc_1y)

max_n225_1d = calc_max(logreturn_n225_1d)
max_n225_1w = calc_max(logreturn_n225_1w)
max_n225_1m = calc_max(logreturn_n225_1m)
max_n225_1y = calc_max(logreturn_n225_1y)

max_nya_1d = calc_max(logreturn_nya_1d)
max_nya_1w = calc_max(logreturn_nya_1w)
max_nya_1m = calc_max(logreturn_nya_1m)
max_nya_1y = calc_max(logreturn_nya_1y)

max_ixic_1d = calc_max(logreturn_ixic_1d)
max_ixic_1w = calc_max(logreturn_ixic_1w)
max_ixic_1m = calc_max(logreturn_ixic_1m)
max_ixic_1y = calc_max(logreturn_ixic_1y)

# Calculating the minimum for the daily, weekly, monthly and yearly databases
min_djia_1d = calc_min(logreturn_djia_1d)
min_djia_1w = calc_min(logreturn_djia_1w)
min_djia_1m = calc_min(logreturn_djia_1m)
min_djia_1y = calc_min(logreturn_djia_1y)

min_gspc_1d = calc_min(logreturn_gspc_1d)
min_gspc_1w = calc_min(logreturn_gspc_1w)
min_gspc_1m = calc_min(logreturn_gspc_1m)
min_gspc_1y = calc_min(logreturn_gspc_1y)

min_n225_1d = calc_min(logreturn_n225_1d)
min_n225_1w = calc_min(logreturn_n225_1w)
min_n225_1m = calc_min(logreturn_n225_1m)
min_n225_1y = calc_min(logreturn_n225_1y)

min_nya_1d = calc_min(logreturn_nya_1d)
min_nya_1w = calc_min(logreturn_nya_1w)
min_nya_1m = calc_min(logreturn_nya_1m)
min_nya_1y = calc_min(logreturn_nya_1y)

min_ixic_1d = calc_min(logreturn_ixic_1d)
min_ixic_1w = calc_min(logreturn_ixic_1w)
min_ixic_1m = calc_min(logreturn_ixic_1m)
min_ixic_1y = calc_min(logreturn_ixic_1y)

# Calculating the skewness for the daily, weekly, mothly and yearly databases
skewness_djia_1d = calc_skewness(logreturn_djia_1d)
skewness_djia_1w = calc_skewness(logreturn_djia_1w)
skewness_djia_1m = calc_skewness(logreturn_djia_1m)
skewness_djia_1y = calc_skewness(logreturn_djia_1y)

skewness_gspc_1d = calc_skewness(logreturn_gspc_1d)
skewness_gspc_1w = calc_skewness(logreturn_gspc_1w)
skewness_gspc_1m = calc_skewness(logreturn_gspc_1m)
skewness_gspc_1y = calc_skewness(logreturn_gspc_1y)

skewness_n225_1d = calc_skewness(logreturn_n225_1d)
skewness_n225_1w = calc_skewness(logreturn_n225_1w)
skewness_n225_1m = calc_skewness(logreturn_n225_1m)
skewness_n225_1y = calc_skewness(logreturn_n225_1y)

skewness_nya_1d = calc_skewness(logreturn_nya_1d)
skewness_nya_1w = calc_skewness(logreturn_nya_1w)
skewness_nya_1m = calc_skewness(logreturn_nya_1m)
skewness_nya_1y = calc_skewness(logreturn_nya_1y)

skewness_ixic_1d = calc_skewness(logreturn_ixic_1d)
skewness_ixic_1w = calc_skewness(logreturn_ixic_1w)
skewness_ixic_1m = calc_skewness(logreturn_ixic_1m)
skewness_ixic_1y = calc_skewness(logreturn_ixic_1y)

# Calculating the kurtosis for the daily, weekly, mothly and yearly databases
kurtosis_djia_1d = calc_kurtosis(logreturn_djia_1d)
kurtosis_djia_1w = calc_kurtosis(logreturn_djia_1w)
kurtosis_djia_1m = calc_kurtosis(logreturn_djia_1m)
kurtosis_djia_1y = calc_kurtosis(logreturn_djia_1y)

kurtosis_gspc_1d = calc_kurtosis(logreturn_gspc_1d)
kurtosis_gspc_1w = calc_kurtosis(logreturn_gspc_1w)
kurtosis_gspc_1m = calc_kurtosis(logreturn_gspc_1m)
kurtosis_gspc_1y = calc_kurtosis(logreturn_gspc_1y)

kurtosis_n225_1d = calc_kurtosis(logreturn_n225_1d)
kurtosis_n225_1w = calc_kurtosis(logreturn_n225_1w)
kurtosis_n225_1m = calc_kurtosis(logreturn_n225_1m)
kurtosis_n225_1y = calc_kurtosis(logreturn_n225_1y)

kurtosis_nya_1d = calc_kurtosis(logreturn_nya_1d)
kurtosis_nya_1w = calc_kurtosis(logreturn_nya_1w)
kurtosis_nya_1m = calc_kurtosis(logreturn_nya_1m)
kurtosis_nya_1y = calc_kurtosis(logreturn_nya_1y)

kurtosis_ixic_1d = calc_kurtosis(logreturn_ixic_1d)
kurtosis_ixic_1w = calc_kurtosis(logreturn_ixic_1w)
kurtosis_ixic_1m = calc_kurtosis(logreturn_ixic_1m)
kurtosis_ixic_1y = calc_kurtosis(logreturn_ixic_1y)

# Calculating the MAD for the daily, weekly, mothly and yearly databases
mad_djia_1d = calc_mad(logreturn_djia_1d)
mad_djia_1w = calc_mad(logreturn_djia_1w)
mad_djia_1m = calc_mad(logreturn_djia_1m)
mad_djia_1y = calc_mad(logreturn_djia_1y)

mad_gspc_1d = calc_mad(logreturn_gspc_1d)
mad_gspc_1w = calc_mad(logreturn_gspc_1w)
mad_gspc_1m = calc_mad(logreturn_gspc_1m)
mad_gspc_1y = calc_mad(logreturn_gspc_1y)

mad_n225_1d = calc_mad(logreturn_n225_1d)
mad_n225_1w = calc_mad(logreturn_n225_1w)
mad_n225_1m = calc_mad(logreturn_n225_1m)
mad_n225_1y = calc_mad(logreturn_n225_1y)

mad_nya_1d = calc_mad(logreturn_nya_1d)
mad_nya_1w = calc_mad(logreturn_nya_1w)
mad_nya_1m = calc_mad(logreturn_nya_1m)
mad_nya_1y = calc_mad(logreturn_nya_1y)

mad_ixic_1d = calc_mad(logreturn_ixic_1d)
mad_ixic_1w = calc_mad(logreturn_ixic_1w)
mad_ixic_1m = calc_mad(logreturn_ixic_1m)
mad_ixic_1y = calc_mad(logreturn_ixic_1y)

# Calculating the IQR for the daily, weekly, mothly and yearly databases
iqr_djia_1d = calc_iqr(logreturn_djia_1d)
iqr_djia_1w = calc_iqr(logreturn_djia_1w)
iqr_djia_1m = calc_iqr(logreturn_djia_1m)
iqr_djia_1y = calc_iqr(logreturn_djia_1y)

iqr_gspc_1d = calc_iqr(logreturn_gspc_1d)
iqr_gspc_1w = calc_iqr(logreturn_gspc_1w)
iqr_gspc_1m = calc_iqr(logreturn_gspc_1m)
iqr_gspc_1y = calc_iqr(logreturn_gspc_1y)

iqr_n225_1d = calc_iqr(logreturn_n225_1d)
iqr_n225_1w = calc_iqr(logreturn_n225_1w)
iqr_n225_1m = calc_iqr(logreturn_n225_1m)
iqr_n225_1y = calc_iqr(logreturn_n225_1y)

iqr_nya_1d = calc_iqr(logreturn_nya_1d)
iqr_nya_1w = calc_iqr(logreturn_nya_1w)
iqr_nya_1m = calc_iqr(logreturn_nya_1m)
iqr_nya_1y = calc_iqr(logreturn_nya_1y)

iqr_ixic_1d = calc_iqr(logreturn_ixic_1d)
iqr_ixic_1w = calc_iqr(logreturn_ixic_1w)
iqr_ixic_1m = calc_iqr(logreturn_ixic_1m)
iqr_ixic_1y = calc_iqr(logreturn_ixic_1y)

# Calculating the Normality Test for the daily, weekly, mothly and yearly databases
normality_djia_1d = normality_test(logreturn_djia_1d)
normality_djia_1w = normality_test(logreturn_djia_1w)
normality_djia_1m = normality_test(logreturn_djia_1m)
normality_djia_1y = normality_test(logreturn_djia_1y)

normality_gspc_1d = normality_test(logreturn_gspc_1d)
normality_gspc_1w = normality_test(logreturn_gspc_1w)
normality_gspc_1m = normality_test(logreturn_gspc_1m)
normality_gspc_1y = normality_test(logreturn_gspc_1y)

normality_n225_1d = normality_test(logreturn_n225_1d)
normality_n225_1w = normality_test(logreturn_n225_1w)
normality_n225_1m = normality_test(logreturn_n225_1m)
normality_n225_1y = normality_test(logreturn_n225_1y)

normality_nya_1d = normality_test(logreturn_nya_1d)
normality_nya_1w = normality_test(logreturn_nya_1w)
normality_nya_1m = normality_test(logreturn_nya_1m)
normality_nya_1y = normality_test(logreturn_nya_1y)

normality_ixic_1d = normality_test(logreturn_ixic_1d)
normality_ixic_1w = normality_test(logreturn_ixic_1w)
normality_ixic_1m = normality_test(logreturn_ixic_1m)
normality_ixic_1y = normality_test(logreturn_ixic_1y)

# Building the analysis vectors with all the parameters calculated so far
median_list1 = [median_djia_1d[0], median_djia_1w[0], median_djia_1m[0], median_djia_1y[0],
                median_gspc_1d[0], median_gspc_1w[0], median_gspc_1m[0], median_gspc_1y[0],
                median_n225_1d[0], median_n225_1w[0], median_n225_1m[0], median_n225_1y[0],
                median_nya_1d[0], median_nya_1w[0], median_nya_1m[0], median_nya_1y[0],
                median_ixic_1d[0], median_ixic_1w[0], median_ixic_1m[0], median_ixic_1y[0]]

median_list2 = [median_djia_1d[1], median_djia_1w[1], median_djia_1m[1], median_djia_1y[1],
                median_gspc_1d[1], median_gspc_1w[1], median_gspc_1m[1], median_gspc_1y[1],
                median_n225_1d[1], median_n225_1w[1], median_n225_1m[1], median_n225_1y[1],
                median_nya_1d[1], median_nya_1w[1], median_nya_1m[1], median_nya_1y[1],
                median_ixic_1d[1], median_ixic_1w[1], median_ixic_1m[1], median_ixic_1y[1]]

median_list = [median_list1, median_list2]

mean_list1 = [mean_djia_1d[0], mean_djia_1w[0], mean_djia_1m[0], mean_djia_1y[0],
              mean_gspc_1d[0], mean_gspc_1w[0], mean_gspc_1m[0], mean_gspc_1y[0],
              mean_n225_1d[0], mean_n225_1w[0], mean_n225_1m[0], mean_n225_1y[0],
              mean_nya_1d[0], mean_nya_1w[0], mean_nya_1m[0], mean_nya_1y[0],
              mean_ixic_1d[0], mean_ixic_1w[0], mean_ixic_1m[0], mean_ixic_1y[0]]

mean_list2 = [mean_djia_1d[1], mean_djia_1w[1], mean_djia_1m[1], mean_djia_1y[1],
              mean_gspc_1d[1], mean_gspc_1w[1], mean_gspc_1m[1], mean_gspc_1y[1],
              mean_n225_1d[1], mean_n225_1w[1], mean_n225_1m[1], mean_n225_1y[1],
              mean_nya_1d[1], mean_nya_1w[1], mean_nya_1m[1], mean_nya_1y[1],
              mean_ixic_1d[1], mean_ixic_1w[1], mean_ixic_1m[1], mean_ixic_1y[1]]
             
mean_list = [mean_list1, mean_list2]

stddev_list1 = [stddev_djia_1d[0], stddev_djia_1w[0], stddev_djia_1m[0], stddev_djia_1y[0],
                stddev_gspc_1d[0], stddev_gspc_1w[0], stddev_gspc_1m[0], stddev_gspc_1y[0],
                stddev_n225_1d[0], stddev_n225_1w[0], stddev_n225_1m[0], stddev_n225_1y[0],
                stddev_nya_1d[0], stddev_nya_1w[0], stddev_nya_1m[0], stddev_nya_1y[0],
                stddev_ixic_1d[0], stddev_ixic_1w[0], stddev_ixic_1m[0], stddev_ixic_1y[0]]

stddev_list2 = [stddev_djia_1d[1], stddev_djia_1w[1], stddev_djia_1m[1], stddev_djia_1y[1],
                stddev_gspc_1d[1], stddev_gspc_1w[1], stddev_gspc_1m[1], stddev_gspc_1y[1],
                stddev_n225_1d[1], stddev_n225_1w[1], stddev_n225_1m[1], stddev_n225_1y[1],
                stddev_nya_1d[1], stddev_nya_1w[1], stddev_nya_1m[1], stddev_nya_1y[1],
                stddev_ixic_1d[1], stddev_ixic_1w[1], stddev_ixic_1m[1], stddev_ixic_1y[1]]

stddev_list = [stddev_list1, stddev_list2]

max_list1 = [max_djia_1d[0], max_djia_1w[0], max_djia_1m[0], max_djia_1y[0],
             max_gspc_1d[0], max_gspc_1w[0], max_gspc_1m[0], max_gspc_1y[0],
             max_n225_1d[0], max_n225_1w[0], max_n225_1m[0], max_n225_1y[0],
             max_nya_1d[0], max_nya_1w[0], max_nya_1m[0], max_nya_1y[0],
             max_ixic_1d[0], max_ixic_1w[0], max_ixic_1m[0], max_ixic_1y[0]]

max_list2 = [max_djia_1d[1], max_djia_1w[1], max_djia_1m[1], max_djia_1y[1],
             max_gspc_1d[1], max_gspc_1w[1], max_gspc_1m[1], max_gspc_1y[1],
             max_n225_1d[1], max_n225_1w[1], max_n225_1m[1], max_n225_1y[1],
             max_nya_1d[1], max_nya_1w[1], max_nya_1m[1], max_nya_1y[1],
             max_ixic_1d[1], max_ixic_1w[1], max_ixic_1m[1], max_ixic_1y[1]]

max_list = [max_list1, max_list2]

min_list1 = [min_djia_1d[0], min_djia_1w[0], min_djia_1m[0], min_djia_1y[0],
             min_gspc_1d[0], min_gspc_1w[0], min_gspc_1m[0], min_gspc_1y[0],
             min_n225_1d[0], min_n225_1w[0], min_n225_1m[0], min_n225_1y[0],
             min_nya_1d[0], min_nya_1w[0], min_nya_1m[0], min_nya_1y[0],
             min_ixic_1d[0], min_ixic_1w[0], min_ixic_1m[0], min_ixic_1y[0]]

min_list2 = [min_djia_1d[1], min_djia_1w[1], min_djia_1m[1], min_djia_1y[1],
             min_gspc_1d[1], min_gspc_1w[1], min_gspc_1m[1], min_gspc_1y[1],
             min_n225_1d[1], min_n225_1w[1], min_n225_1m[1], min_n225_1y[1],
             min_nya_1d[1], min_nya_1w[1], min_nya_1m[1], min_nya_1y[1],
             min_ixic_1d[1], min_ixic_1w[1], min_ixic_1m[1], min_ixic_1y[1]]

min_list = [min_list1, min_list2]

skewness_list1 = [skewness_djia_1d[0], skewness_djia_1w[0], skewness_djia_1m[0], skewness_djia_1y[0],
                  skewness_gspc_1d[0], skewness_gspc_1w[0], skewness_gspc_1m[0], skewness_gspc_1y[0],
                  skewness_n225_1d[0], skewness_n225_1w[0], skewness_n225_1m[0], skewness_n225_1y[0],
                  skewness_nya_1d[0], skewness_nya_1w[0], skewness_nya_1m[0], skewness_nya_1y[0],
                  skewness_ixic_1d[0], skewness_ixic_1w[0], skewness_ixic_1m[0], skewness_ixic_1y[0]]

skewness_list2 = [skewness_djia_1d[1], skewness_djia_1w[1], skewness_djia_1m[1], skewness_djia_1y[1],
                  skewness_gspc_1d[1], skewness_gspc_1w[1], skewness_gspc_1m[1], skewness_gspc_1y[1],
                  skewness_n225_1d[1], skewness_n225_1w[1], skewness_n225_1m[1], skewness_n225_1y[1],
                  skewness_nya_1d[1], skewness_nya_1w[1], skewness_nya_1m[1], skewness_nya_1y[1],
                  skewness_ixic_1d[1], skewness_ixic_1w[1], skewness_ixic_1m[1], skewness_ixic_1y[1]]

skewness_list = [skewness_list1, skewness_list2]

kurtosis_list1 = [kurtosis_djia_1d[0], kurtosis_djia_1w[0], kurtosis_djia_1m[0], kurtosis_djia_1y[0],
                  kurtosis_gspc_1d[0], kurtosis_gspc_1w[0], kurtosis_gspc_1m[0], kurtosis_gspc_1y[0],
                  kurtosis_n225_1d[0], kurtosis_n225_1w[0], kurtosis_n225_1m[0], kurtosis_n225_1y[0],
                  kurtosis_nya_1d[0], kurtosis_nya_1w[0], kurtosis_nya_1m[0], kurtosis_nya_1y[0],
                  kurtosis_ixic_1d[0], kurtosis_ixic_1w[0], kurtosis_ixic_1m[0], kurtosis_ixic_1y[0]]

kurtosis_list2 = [kurtosis_djia_1d[1], kurtosis_djia_1w[1], kurtosis_djia_1m[1], kurtosis_djia_1y[1],
                  kurtosis_gspc_1d[1], kurtosis_gspc_1w[1], kurtosis_gspc_1m[1], kurtosis_gspc_1y[1],
                  kurtosis_n225_1d[1], kurtosis_n225_1w[1], kurtosis_n225_1m[1], kurtosis_n225_1y[1],
                  kurtosis_nya_1d[1], kurtosis_nya_1w[1], kurtosis_nya_1m[1], kurtosis_nya_1y[1],
                  kurtosis_ixic_1d[1], kurtosis_ixic_1w[1], kurtosis_ixic_1m[1], kurtosis_ixic_1y[1]]

kurtosis_list = [kurtosis_list1, kurtosis_list2]

mad_list1 = [mad_djia_1d[0], mad_djia_1w[0], mad_djia_1m[0], mad_djia_1y[0],
             mad_gspc_1d[0], mad_gspc_1w[0], mad_gspc_1m[0], mad_gspc_1y[0],
             mad_n225_1d[0], mad_n225_1w[0], mad_n225_1m[0], mad_n225_1y[0],
             mad_nya_1d[0], mad_nya_1w[0], mad_nya_1m[0], mad_nya_1y[0],
             mad_ixic_1d[0], mad_ixic_1w[0], mad_ixic_1m[0], mad_ixic_1y[0]]

mad_list2 = [mad_djia_1d[1], mad_djia_1w[1], mad_djia_1m[1], mad_djia_1y[1],
             mad_gspc_1d[1], mad_gspc_1w[1], mad_gspc_1m[1], mad_gspc_1y[1],
             mad_n225_1d[1], mad_n225_1w[1], mad_n225_1m[1], mad_n225_1y[1],
             mad_nya_1d[1], mad_nya_1w[1], mad_nya_1m[1], mad_nya_1y[1],
             mad_ixic_1d[1], mad_ixic_1w[1], mad_ixic_1m[1], mad_ixic_1y[1]]

mad_list = [mad_list1, mad_list2]

iqr_list1 = [iqr_djia_1d[0], iqr_djia_1w[0], iqr_djia_1m[0], iqr_djia_1y[0],
             iqr_gspc_1d[0], iqr_gspc_1w[0], iqr_gspc_1m[0], iqr_gspc_1y[0],
             iqr_n225_1d[0], iqr_n225_1w[0], iqr_n225_1m[0], iqr_n225_1y[0],
             iqr_nya_1d[0], iqr_nya_1w[0], iqr_nya_1m[0], iqr_nya_1y[0],
             iqr_ixic_1d[0], iqr_ixic_1w[0], iqr_ixic_1m[0], iqr_ixic_1y[0]]

iqr_list2 = [iqr_djia_1d[1], iqr_djia_1w[1], iqr_djia_1m[1], iqr_djia_1y[1],
             iqr_gspc_1d[1], iqr_gspc_1w[1], iqr_gspc_1m[1], iqr_gspc_1y[1],
             iqr_n225_1d[1], iqr_n225_1w[1], iqr_n225_1m[1], iqr_n225_1y[1],
             iqr_nya_1d[1], iqr_nya_1w[1], iqr_nya_1m[1], iqr_nya_1y[1],
             iqr_ixic_1d[1], iqr_ixic_1w[1], iqr_ixic_1m[1], iqr_ixic_1y[1]]

iqr_list = [iqr_list1, iqr_list2]

jb_list = [normality_djia_1d[0], normality_djia_1w[0], normality_djia_1m[0], normality_djia_1y[0],
           normality_gspc_1d[0], normality_gspc_1w[0], normality_gspc_1m[0], normality_gspc_1y[0],
           normality_n225_1d[0], normality_n225_1w[0], normality_n225_1m[0], normality_n225_1y[0],
           normality_nya_1d[0], normality_nya_1w[0], normality_nya_1m[0], normality_nya_1y[0],
           normality_ixic_1d[0], normality_ixic_1w[0], normality_ixic_1m[0], normality_ixic_1y[0]]

shapiro_list = [normality_djia_1d[1], normality_djia_1w[1], normality_djia_1m[1], normality_djia_1y[1],
                normality_gspc_1d[1], normality_gspc_1w[1], normality_gspc_1m[1], normality_gspc_1y[1],
                normality_n225_1d[1], normality_n225_1w[1], normality_n225_1m[1], normality_n225_1y[1],
                normality_nya_1d[1], normality_nya_1w[1], normality_nya_1m[1], normality_nya_1y[1],
                normality_ixic_1d[1], normality_ixic_1w[1], normality_ixic_1m[1], normality_ixic_1y[1]]

# anderson_list = [normality_djia_1d[2], normality_djia_1w[2], normality_djia_1m[2], normality_djia_1y[2],
#                  normality_gspc_1d[2], normality_gspc_1w[2], normality_gspc_1m[2], normality_gspc_1y[2],
#                  normality_n225_1d[2], normality_n225_1w[2], normality_n225_1m[2], normality_n225_1y[2],
#                  normality_nya_1d[2], normality_nya_1w[2], normality_nya_1m[2], normality_nya_1y[2],
#                  normality_ixic_1d[2], normality_ixic_1w[2], normality_ixic_1m[2], normality_ixic_1y[2]]

ks_list = [normality_djia_1d[2], normality_djia_1w[2], normality_djia_1m[2], normality_djia_1y[2],
           normality_gspc_1d[2], normality_gspc_1w[2], normality_gspc_1m[2], normality_gspc_1y[2],
           normality_n225_1d[2], normality_n225_1w[2], normality_n225_1m[2], normality_n225_1y[2],
           normality_nya_1d[2], normality_nya_1w[2], normality_nya_1m[2], normality_nya_1y[2],
           normality_ixic_1d[2], normality_ixic_1w[2], normality_ixic_1m[2], normality_ixic_1y[2]]

df = construct_dataframe(median_list, mean_list, stddev_list, max_list, min_list, skewness_list, kurtosis_list, mad_list, iqr_list, jb_list, shapiro_list, ks_list)

directory = "C:/Users/henri/OneDrive/Mestrado Engenharia Financeira/Segundo Semestre/Estratégia de Trading/Monitorias/Monitoria 1/databases/"
df.to_csv(directory + 'FinalResults.csv')
```
